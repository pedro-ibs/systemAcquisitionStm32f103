```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./1khz.csv')


df['TEMPO_US'] = range(0, 7680);
df['TEMPO_US'] = df['TEMPO_US'].apply(lambda x: x/51200)
df['AMOSTRA'] = df['AMOSTRA'].apply(lambda x: 3.3*(x/4096))

fig, axs = plt.subplots(2, figsize=(30,14) )
fig.suptitle('Medidas de 1kHZ')

axs[0].set_ylabel('Tensão [ V ]')
axs[1].set_ylabel('Tensão [ V ]')
axs[1].set_xlabel('Tempo [ S ]')


axs[0].plot(df['TEMPO_US'], df['AMOSTRA'], 'b')
axs[1].plot(df['TEMPO_US'], df['AMOSTRA'], 'b')

axs[1].set_ylim([-0.1, 2])
axs[1].set_xlim([0.00, 0.005])

plt.savefig('Medidas1kHZ.png')

plt.show()

```
